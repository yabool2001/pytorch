```python
import numpy as np
import torch
```

```python
device = torch.accelerator.current_accelerator().type if torch.accelerator.is_available() else "cpu"
print ( f"{device=}" )
print ( f"{torch.__version__=}" )
```

```python
# torch.empty creates a tensor without initializing the values (they will be whatever is in memory at the time)
scalar = torch.empty ( size = ( 1 , ) ) ; print ( f"{scalar=}" )
# torch.rand creates a tensor with random values between 0 and 1
vector = torch.rand ( size = ( 2 , ) ) ; print ( f"{vector=}" )
# torch.zeros creates a tensor filled with zeros
matrix = torch.zeros ( size = ( 2 , 3 ) ) ; print ( f"{matrix=}" )
# torch.ones creates a tensor filled with ones
tensor = torch.ones ( size = ( 2 , 3 , 4 ) , requires_grad = True ) ; print ( f"{tensor=}" ) # 3 dimensions
# Alternatively, we can have it all simpler without the size, e.g.:
tensor2 = torch.empty ( 2, 3, 4 , requires_grad = True ) ; print ( f"{tensor2=}" )
```

```python
print ( f"{tensor.shape=}" ) ; print ( f"{tensor.size()=}" ) # shape and size are the same
print ( f"{tensor.ndim=}" )
print ( f"{tensor.dtype=}" )
print ( f"{tensor.device=}" )
print ( f"{tensor.requires_grad=}" )
```

```python
tensor3 = tensor + tensor2 ; print ( f"{tensor3=}" )
 # in-place addition. It's faster and uses less memory, but it can cause issues with autograd if not used carefully. Use with caution.
tensor3.add_( tensor ) ; print ( f"{tensor3=}" )
```

```python
print ( f"{matrix[ : , 0 ]=}" ) # all rows, first column (column index starts at 0)
print ( f"{matrix[ 0 , 0 ]=}" ) # first row, first column
print ( f"{matrix[ 0 , 0 ].item()=}" ) # get the value as a standard Python number  (scalar)
```

```python
a = np.empty ( ( 2 , 3 ) ) ; print ( f"{a=}" )
b = torch.from_numpy ( a ) ; print ( f"{b=}" )
c = torch.from_numpy ( a ).to ( dtype = torch.float32 ) ; print ( f"{c=}" )
print ( f"{a.dtype=} {a.shape=}, {b.dtype=} {b.shape=}, {c.dtype=} {c.shape=}" )
```
