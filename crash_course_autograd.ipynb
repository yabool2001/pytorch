```python
import torch

device = torch.accelerator.current_accelerator () .type if torch.accelerator.is_available () else "cpu"
print ( f"Using {device} device" )
print ( torch.__version__ )
```

```python
vector1 = torch.rand ( 3 , requires_grad = True ) ; print ( f"{vector1=}" )
y = vector1 + 2 ; print ( f"{y=}" )
z = y * y * 2 ; print ( f"{z=}" )
z = z.mean() ; print ( f"{z=}" )
z.backward () ; print ( f"{vector1.grad=}" ) # dz/d(vector1) which is the gradient of z with respect to vector1
```
