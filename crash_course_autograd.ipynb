```python
import torch

device = torch.accelerator.current_accelerator () .type if torch.accelerator.is_available () else "cpu"
print ( f"Using {device} device" )
print ( torch.__version__ )
```

```python
vector1 = torch.tensor ( [ 1. , 2. , 3. ] , requires_grad = True ) ; print ( f"{vector1=}" )
y = vector1 * 2 ; print ( f"{y=} {y.grad_fn=}" )
y = y.mean () ; print ( f"{y=}" )
y.backward () ; print ( f"{vector1.grad=}" ) # dy/d(vector1) which is the gradient of y with respect to vector1
```

```python
print ( f"{vector1.requires_grad=}" )
with torch.no_grad () :
    vector2 = vector1 * 3 ; print ( f"{vector2=} {vector2.grad_fn=}" )
print ( f"{vector2.requires_grad=}" )
```
