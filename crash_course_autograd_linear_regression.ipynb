```python
import torch

device = torch.accelerator.current_accelerator () .type if torch.accelerator.is_available () else "cpu"
print ( f"Using {device} device" )
print ( torch.__version__ )
```

```python
X = torch.tensor ( [ [ 1. , 2. , 3. , 4. , 5. , 6. , 7. , 8. ] ] , dtype = torch.float32 ) ; print ( f"{X=} {X.grad=} {X.is_leaf=}" )
Y = torch.tensor ( [ [ 2. , 4. , 6. , 8. , 10. , 12. , 14. , 16. ] ] , dtype = torch.float32 ) ; print ( f"{Y=} {Y.grad=} {Y.is_leaf=}" )
w = torch.tensor ( 0. , dtype = torch.float32 , requires_grad = True ) ; print ( f"{w=} {w.grad=} {w.is_leaf=}" )

# forward pass (model output)
def forward ( x ) :
    return w * x

# loss = MSE ( mean squared error )
def loss ( y , y_pred ) :
    return ( ( y_pred - y ) ** 2 ) .mean ()

X_test = 5.0
print ( f"Prediction before training: f({X_test}) = {forward ( X_test ) .item () }" )
```
